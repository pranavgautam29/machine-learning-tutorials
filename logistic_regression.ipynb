## **Logistic Regression**

Logistic Regression is a classification algorithm. It is used to predict the outcome of a categorical dependent variable on the based on some independent variables.

Import essential libraries and modules

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

Load the data

```python
FILE = 'heart.csv'
data = pd.read_csv(FILE)
data.head()
```

Basic EDA

```python
data.describe()
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

Define features and target 

```python
data.columns
```

```python
FEATURES = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach','exang', 'oldpeak', 'slope', 'ca', 'thal']
X = data[FEATURES]
y = data.target
X.head()
```

Train Test Split

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,test_size=0.2,random_state=20)
X_train.head()
```

Resize the distribution of values using StandardScaler

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

Define and fit the model

```python
log_reg = LogisticRegression(random_state=1,C=1,fit_intercept=True)
```

```python
log_reg.fit(X_train,y_train)
```

Make predictions

```python
preds = log_reg.predict(X_train)
```

Check the score

```python
log_reg.score(X_train,y_train)
```

```python
log_reg.score(X_test,y_test)
```
