# **Polynomial Regression**

Polynomial Regression is a type of model which predicts the dependent variable when the relationship between the independent and the dependent variable is non-linear.


Equation of Polynomial Regression : 
y = b1 + b2x1 + b3x2^2 + b4x3^3 + ... + bnxn^n

Import the essential libraries and modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

Load the data


```python
FILE = 'Housing.csv'
data = pd.read_csv(FILE)
data.head()
```

Basic Data Explration


```python
data.describe()
```

```python
data.isnull()
```

Define Feature and Target variables

```python
y = data.price
FEATURES = ['area','bedrooms','bathrooms','stories','parking']
X = data[FEATURES]
X.head()
```

Train Test Split

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,test_size=0.2,random_state=10)
X_train.head()
```

Define the Model

```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X_train)
poly.fit(X_poly,y_train)
```

```python
linear_model = LinearRegression()
linear_model.fit(X_poly,y_train)
```

Prediction

```python
preds = linear_model.predict(poly.fit_transform(X_test))
```

Validation

```python
score = mean_absolute_error(preds,y_test)
score
```
